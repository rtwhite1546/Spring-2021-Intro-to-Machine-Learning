```python
import glob
import pandas as pd
```

```python
datafiles = glob.glob('data/PRSA/*')

data = []

for file in datafiles:
    data.append(pd.read_csv(file))
    
data = pd.concat(data)

data
```

```python
data = data.drop(columns = ['No'])
data = data.dropna()

data
```

```python
data = pd.get_dummies(data, columns = ['station'])

data
```
